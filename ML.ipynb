```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
music_data = pd.read_csv('music.csv')
music_data
```

```python
# we need to know make input and output dataset so , we will discard the genre(output) column

X = music_data.drop(columns=['genre'])
X
```

```python
#lets make the output dataset
y = music_data['genre']
y
```

```python
# splitting data on training and test part with 20% for testing

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
#MODEL decision Tree
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
predictions = model.predict(X_test)
print(f'The predictions are:',predictions)
score = accuracy_score(y_test,predictions)
score
```
